```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Mobile Reviews Sentiment.csv")
mean_p = df['price_usd'].mean()
mean_rat = df['rating'].mean()
overpriced = df[(df['price_usd'] > mean_p) & (df['rating'] < mean_rat)]

plt.figure(figsize=(8,6))
plt.scatter(df['price_usd'], df['rating'], color='gray', label='All Models')
plt.scatter(overpriced['price_usd'], overpriced['rating'], color='red', label='Overpriced Models')
plt.axhline(mean_rat, color='blue', linestyle='--', label='Mean Rating')
plt.axvline(mean_p, color='green', linestyle='--', label='Mean Price')
plt.xlabel('Price $')
plt.ylabel('Rating')
plt.title('Overpriced Models (High Price, Low Rating)')
plt.legend()
plt.show()

print(overpriced[['brand', 'model', 'price_usd', 'rating']])
```
