```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("Mobile Reviews Sentiment.csv")
print(df.head())
avg_rating = df.groupby('brand')['rating'].mean().sort_values(ascending=False)
print(avg_rating)

plt.figure(figsize=(8,5))
avg_rating.plot(kind='bar', color='blue')

plt.title('Average Overall Rating by Mobile Brand')
plt.xlabel('Mobile Brand')
plt.ylabel('Average Rating')
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.7)

plt.show()
```
