```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("Mobile Reviews Sentiment.csv")
avg_price = df.groupby('brand')['price_usd'].mean().sort_values(ascending=False)
print(avg_price)
plt.figure(figsize=(8,5))
avg_price.plot(kind='bar', color='orange', edgecolor='black')
plt.title('Average Price (USD) by Mobile Brand')
plt.xlabel('Mobile Brand')
plt.ylabel('Average Price (USD)')
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.7)
for i, val in enumerate(avg_price):
    plt.text(i, val, f"${val:.0f}", ha='center', va='bottom', fontsize=9)
plt.show()
```
