```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Mobile Reviews Sentiment.csv")

plt.figure(figsize=(8,6))
df.boxplot(column='rating', by='sentiment', grid=False, patch_artist=True,
            boxprops=dict(facecolor='blue', color='green'),
            medianprops=dict(color='red'))

plt.title('Sentiment vs Rating Comparison')
plt.suptitle('')
plt.xlabel('Sentiment')
plt.ylabel('Rating')
plt.show()
```
