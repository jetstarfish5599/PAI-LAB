```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Software houses PK .csv")
services = df['Services'].fillna("").str.lower()

industry_keywords = {
    'finance': ['fintech', 'finance', 'bank'],
    'healthcare': ['health', 'hospital', 'clinic'],
    'ecommerce': ['e-commerce', 'shopping', 'retail'],
    'education': ['education', 'school', 'college', 'university'],
    'logistics': ['logistic', 'supply chain','delivery'],
    'real_estate': ['real estate', 'property', 'estate'],
    'telecom': [ 'telecommunication', 'sms'],
    'gaming': ['game', 'gaming', 'unity', 'steam'],
    'manufacturing': ['manufacture', 'industrial'],
    'marketing': ['digital marketing', 'seo', 'advertise'],
    'government': ['government', 'govt', 'public'],
}

def detect_industries(text):
    found = set()
    for ind, kws in industry_keywords.items():
        for kw in kws:
            if kw in text:
                found.add(ind)
                break
    return list(found) if found else ['other']

industry_series = services.apply(detect_industries)
industry_exploded = industry_series.explode().value_counts()

plt.figure(figsize=(8,6))
industry_exploded.plot(kind='bar', color='orange')
plt.title('Inferred Industries from Services')
plt.xlabel('Industry')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.5)
plt.show()
```
